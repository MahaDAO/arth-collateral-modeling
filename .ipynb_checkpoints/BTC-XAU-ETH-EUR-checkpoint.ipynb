Crypto Simulation
===============

This notebook simulates the rebalancing algorithim with a variety of cryptocurrencies, stablecoins and tokenized assets.

The stablecoins used in this simulation are USD stablecoins as the marketcap of other fiat stablecoins are too small to come into consideration. Further, the time horizon choosen from 2019 onwards, since smart-contracts were only introduced in 2015 (with the launch of Ethereum) and tokenzied assets like PAXG (tokenized gold) was onyl available in late 2019

We use the following parameters:

- Stop sell of 40%
- Stop loss of 30%
- 5% slippage for every rebalance
- Rebalance to happen once every 7 days
- 50% of the appreciated value from the appreciating asset, is sold off during every rebalance
- Starting with a buying power of 100 (so that it is easy to measure % increase)

And assets are distributed with the following ratios:

- Euro (50% starting buying power)
- Gold (20% starting buying power) 
- WBTC (20% starting buying power)
- ETH (10% starting buying power)

### Results

At the end of the simulation, the vault has rebalanced itself around 11 times over the course of 1 year and is up by 87%.
