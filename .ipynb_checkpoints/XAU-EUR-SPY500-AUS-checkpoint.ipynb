Forex & Equities Simulation
=====================

This notebook simulates the rebalancing algorithim with a variety of forex currencies, equities and gold.

This simluation uses 20 years of data and accounts for market scenarios like the 2008 financial crisis and the 2020 corona virus epidemic. 

The portfolio contains 

We use the following parameters:

- Stop sell of 40%
- Stop loss of 30%
- 5% slippage for every rebalance
- Rebalance to happen once every 7 days
- 50% of the appreciated value from the appreciating asset, is sold off during every rebalance
- Starting with a buying power of 100 (so that it is easy to measure % increase)

And assets are distributed with the following ratios:

- Gold (50% starting buying power)
- AUD (20% starting buying power) 
- SPY500 ETF (20% starting buying power)
- EUR (10% starting buying power)

### Results

At the end of the simulation, the vault has rebalanced itself around 120 times over the course of 20 years and is up by 334% against the dollar.
