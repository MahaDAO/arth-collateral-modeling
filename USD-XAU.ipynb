Simple XAU-USD Simulation
===============

This notebook simulates the rebalancing algorithim with a simple 50-50 portfolio of USD and Gold. Buying power is measured in Chinese Yuan.

We use the following parameters:

- Stop sell of 40%
- Stop loss of 30%
- 5% slippage for every rebalance
- Rebalance to happen once every 7 days
- 50% of the appreciated value from the appreciating asset, is sold off during every rebalance
- Starting with a buying power of 100 (so that it is easy to measure % increase)

And assets are distributed with the following ratios:

- USD (50% starting buying power
- XAU (50% starting buying power)

### Results

At the end of the simulation, the vault has rebalanced itself around 11 times over the course of 1 year and is up by 87%.
